```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fname = ('/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/sales_predictions_2023.csv')
df = pd.read_csv(fname)
df.head()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(10,5))

sns.histplot(data = df , x ='Item_Weight', ax=axes[0,0])
#ax[0,0].set_title('Count for Item Weight')

sns.histplot(data = df , x ='Item_Outlet_Sales', ax=axes[1,0], color='green')
#ax[1,0].set_title('Count for Item Weight')

sns.histplot(data = df , x ='Item_MRP', ax=axes[1,1], color='purple')
#ax[1,1].set_title('Item_MRP')
sns.histplot(data = df , x ='Outlet_Establishment_Year', ax=axes[0,1], color = 'orange')

fig.suptitle('Count for Numiric features')
fig.tight_layout()
```

```python
fig, axes = plt.subplots(ncols=1, nrows=4, figsize=(7,7))

sns.boxplot(data = df , x ='Item_Weight', ax=axes[0])
#ax[0,0].set_title('Count for Item Weight')

sns.boxplot(data = df , x ='Item_Outlet_Sales', ax=axes[1], color='green')
#ax[1,0].set_title('Count for Item Weight')

sns.boxplot(data = df , x ='Item_MRP', ax=axes[2], color='purple')
#ax[1,1].set_title('Item_MRP')
sns.boxplot(data = df , x ='Outlet_Establishment_Year', ax=axes[3], color = 'orange')

fig.suptitle('Statistcal distribution of numeric features')
#fig.tight_layout()
plt.subplots_adjust(hspace= 1.5)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(10,5))

sns.boxplot(data = df , x ='Item_Weight', ax=axes[0,0])
#ax[0,0].set_title('Count for Item Weight')

sns.boxplot(data = df , x ='Item_Outlet_Sales', ax=axes[1,0], color='green')
#ax[1,0].set_title('Count for Item Weight')

sns.boxplot(data = df , x ='Item_MRP', ax=axes[1,1], color='purple')
#ax[1,1].set_title('Item_MRP')
sns.boxplot(data = df , x ='Outlet_Establishment_Year', ax=axes[0,1], color = 'orange')

fig.suptitle('Statistcal distribution of numeric features')
#fig.tight_layout()
plt.subplots_adjust(wspace=0.5, hspace= 0.6)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(10,5))

sns.countplot(data = df , x ='Item_Fat_Content', ax=axes[0,0])
#ax[0,0].set_title('Count for Item Weight')

sns.countplot(data = df , x ='Item_Type', ax=axes[1,0], color='green')
axes[1,0].set_xticklabels(axes[1,0].get_xticklabels(), rotation=90)
#ax[1,0].set_title('Count for Item Weight')

sns.countplot(data = df , x ='Outlet_Size', ax=axes[1,1], color='purple')
#ax[1,1].set_title('Item_MRP')

sns.countplot(data = df , x ='Outlet_Identifier', ax=axes[0,1], color = 'orange')
axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(), rotation=45)


fig.suptitle('frequency of each class of categorial features')
#fig.tight_layout()
plt.subplots_adjust(wspace=0.5, hspace= 0.6)
```

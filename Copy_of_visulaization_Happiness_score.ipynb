```python
#Which region have highest happiness score, and which have lowest Which regions have countries with extream (outliers) happiness score?


ax= sns.boxplot(x= 'Region', y = 'Happiness Score' , data = df)
plt.xticks(rotation = 90)
ax.set_title('Distribution of Happiness Score by Region')
plt.show()
#Answer:
#1-highest happiness score: westren Eroupe, lowest : sub sahar african , outliers in latin america
```

```python
df['Year']=df['Year'].astype(str)
print(df['Year'].dtype)
```

```python
#Show how the global happiness score change over time

ax= sns.lineplot( x= 'Year' , y= 'Happiness Score' , hue = 'Region' , data = df,  marker='o')
ax.legend(bbox_to_anchor =(1,1),loc='upper left')
ax.set_title('Happiness Score over Time')
```
